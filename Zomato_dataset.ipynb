```python
df_country=pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
df.columns
```

```python
# here we merge two data frame with country code in it and we perform left joint in this merge
pd.merge(df,df_country,on='Country Code' , how='left')
```

```python
# here we assign the above dataset with the new variable name
final_df=pd.merge(df,df_country,on='Country Code' , how='left')

# it will print the first two records if the dataframe
final_df.head(2)
```

```python
# to check the datatype of the previous data frame
final_df.dtypes
```

```python
# print all the columns of the data
final_df.columns
```

```python
# here we basically count how many country are there and how many records are there in that particular country
final_df['Country'].value_counts()
```

```python
# here we get all the country count
final_df.Country.value_counts().index
```

```python
# we put it with variable name
country_names=final_df.Country.value_counts().index
```

```python
# it will print only values of the country
country_val=final_df.Country.value_counts().values
```

```python
## PIE CHART
plt.pie(country_val,labels=country_names)
```

```python
## PIE CHART - Top 3 country which uses zomato
plt.pie(country_val[:3],labels=country_names[:3])
```

```python
plt.pie(country_val[:3],labels=country_names[:3],autopct='%1.2f%%')
```

Observation: zomato maximum records and transactoin are from INDIA after that from UNITED STATES and then from UNITED KINGDOM

```python
final_df.columns
```

```python
final_df.groupby(['Aggregate rating','Rating color','Rating text']).size()
```

```python
# here we just group in three columns of the data frame
final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index()
```

```python
# Just rename the column 0 to rating count
final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
# we load the above data in a new variable i.e."ratings"
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
# print the ratings
ratings
```

# Observation:
1. whenever the rating is between  4.5 to 4.9---> Excellent
2. when the rating are between 4.0 to 4.4---> Very good
3. when the rating is between 3.5 to 3.9---> good
4. when the rating is between 3.0 to 3.4---> AVerage
5. when the rating is between 2.5 ot 2.9---> Average
6. When the rating is between 2.0 to 2.4---> Poor  

```python
# x axis--> Aggreagate rating and y axis--> Rating count
sns.barplot(x='Aggregate rating',y='Rating Count',data=ratings)

# basically we just increase the size of the graph with hellp of matplotlib
# rcPArms--> it is a type of parameter which you want to increase
plt.matplotlib.rcParams['figure.figsize']=(15,7)
```

```python
# hue-->> is basically used to assign colors
sns.barplot(x='Aggregate rating',y='Rating Count', hue='Rating color',data=ratings)
```

```python
# palette -->> is basically used to transform the colors into its original formats
# we can also assign the colors according to u in this palette
sns.barplot(x='Aggregate rating',y='Rating Count', hue='Rating color',data=ratings, palette=['white','red','orange','yellow','green','green'])
```

# Observation:
1. Not rated count is very high
2. MAxium number of rating are between 2.5 to 3.4

```python
# count plot
sns.countplot(x='Rating color',data=ratings, palette=['white','red','orange','yellow','green','green'])
```

```python
# count plot
sns.countplot(x='Aggregate rating',data=ratings, palette=['white','red','orange','yellow','green','green'])
```
